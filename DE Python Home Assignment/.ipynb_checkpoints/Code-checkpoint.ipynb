```python
import pandas as pd
import re

df = pd.read_excel(r'sample physician information.xlsx')

info = []

for ind in df.index:
    info.append(df['Fulll Information'][ind])
    


cpso = []
names = []
phone_num = []


# Extracting CPSO#

for x in range(len(info)):
    m.append(re.findall('CPSO#', info[x]))
    if m[x] == '':
        m[x] = False
    else:
        m[x] = True

for x in range(len(info)):
    if  m[x] == True:
        test = re.findall(r'\d+', info[x])
        cpso.append(test[0])
    else:
        cpso.append(None)
        

# Phone numbers 

m = []
phone_start = []
phone_end = []
phone = []

for x in range(len(info)):
    m.append(re.search('Phone:', info[x]))
    

for x in range(len(m)):  
    phone_start.append(m[x].end())


for x in range(len(info)):
    if 'Fax' in info[x]:
        phone_end.append(info[x].find('Fax'))
    elif 'This' in info[x]:
        phone_end.append(info[x].find('This'))
    else:
        phone_end.append(info[x].find('Area'))

        
for x in range(len(info)):
    if phone_end[x] == -1:
        temp = info[x][phone_start[x]:]
    else:
        temp = info[x][phone_start[x]:phone_end[x]]
    
    phone.append(re.findall(r'\d+', temp))
    if len(phone[x]) == 3:
        phone_num.append(phone[x][0]+phone[x][1]+phone[x][2])
    else:
        phone_num.append(phone[x][0])
print(phone_num)


# Extracting Names

name_end = []

for x in range(len(info)):
        name_end.append(info[x].find(' (CPSO#'))

for x in range(len(info)):
        names.append(info[x][:name_end[x]])

# Compling them into one place

info_check = {
        'CPSO#': cpso,
        'Name': names,
        'Phone Number': phone_num
               }

phy_info = pd.DataFrame(info_check)

print(phy_info)


phy_info.to_excel('phy_info.xlsx')

print('DataFrame is exported to Excel File successfully.')


    
```
